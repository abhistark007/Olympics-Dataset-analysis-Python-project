# 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df
```

```python
data=[]
for i in df["Sport"].unique():
    data.append([i,len(df[(df['Sport']==i) & (df['Medal']=="Gold")])])
data
pd.DataFrame(data,columns = ['Sport','GoldMedal']).sort_values(by='GoldMedal', ascending=False).head().plot(x = 'Sport', y = 'GoldMedal', kind = 'bar', figsize = (5,5))
```

```python
# Aquatics has most number of gold medals so far
```

# 6. In which year India won first Gold Medal in Summer Olympics?

# 8. Which sport is having most female Gold Medalists? (Top 5)

```python
df
```

```python
data=[]
for i in df["Sport"].unique():
    data.append([i,len(df[(df['Sport']==i) & (df["Medal"]=="Gold") & (df["Gender"]=="Women")])])
data
```

```python
pd.DataFrame(data,columns = ['Sport','number of Female gold medalist']).sort_values(by='number of Female gold medalist', ascending=False).head().plot(x = 'Sport', y = 'number of Female gold medalist', kind = 'bar', figsize = (5,5))
```
